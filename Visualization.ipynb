```python
plt.figure(figsize=(10,5))
np.random.seed(123)

# Generate a random float number using a numpy between any float range
wt_m = np.random.uniform(60.1,68.5, size = 10)
ht_m = np.random.uniform(124.3,168.7, size = 10)

wt_f = np.random.uniform(45.5,61.5, size = 10)
ht_f = np.random.uniform(110.3,155.5, size = 10)
 
plt.scatter(wt_m,ht_m,c='red',marker='s',label = 'Male' )
plt.scatter(wt_f,ht_f,c='green',marker='*', s = 100 * 2, label = 'Female')
plt.xlabel('weight', fontsize=16)
plt.ylabel('height', fontsize=16)
plt.title('scatter plot - height vs weight',fontsize=20)
plt.legend()
plt.show()
```

### Bar Chart

```python
City=['Delhi','Beijing','Washington','Tokyo','Moscow']

Price_sqft_in_K=[60,40,70,65,45]
 
plt.barh(City,Price_sqft_in_K,color='skyblue',edgecolor='yellow')

plt.xlabel('Price per SQ FT', fontsize=16)
plt.ylabel('City', fontsize=16)
plt.title('Barchart - Price_sft_in_K across cities',fontsize=20)
plt.show()
```

```python
City=['Delhi','Beijing','Washington','Tokyo','Moscow']

Price_sqft_in_K_center = [60,40,70,65,45]
Price_sqft_in_K_outer = [45,22,50,52,20]

location = ['center', 'outer']
 
plt.bar(City,Price_sqft_in_K_center,color='skyblue',edgecolor='black')
plt.bar(City,Price_sqft_in_K_outer, bottom=Price_sqft_in_K_center, color='darkblue',edgecolor='black')

plt.xlabel('City', fontsize=16)
plt.ylabel('Price_sft_in_K', fontsize=16)
plt.title('Barchart - Price_sft_in_K across cities',fontsize=20)
plt.legend(location, loc = 0)
plt.show()
```

## Box plot

Description
A Box and Whisker Plot (or Box Plot) is a convenient way of visually displaying the data distribution through their quartiles.

The lines extending parallel from the boxes are known as the “whiskers”, which are used to indicate variability outside the upper and lower quartiles. Outliers are sometimes plotted as individual dots that are in-line with whiskers. Box Plots can be drawn either vertically or horizontally.

Although Box Plots may seem primitive in comparison to a Histogram or Density Plot, they have the advantage of taking up less space, which is useful when comparing distributions between many groups or datasets.

For some distributions/datasets, you will find that you need more information than the measures of central tendency (median, mean, and mode)

- Boxplots are a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).
- median (Q2/50th Percentile): the middle value of the dataset.
- first quartile (Q1/25th Percentile): the middle number between the smallest number (not the “minimum”) and the median of the dataset.
- third quartile (Q3/75th Percentile): the middle value between the median and the highest value (not the “maximum”) of the dataset.
- interquartile range (IQR): 25th to the 75th percentile.
- whiskers (shown in blue)
- outliers (shown as green circles)
    “maximum”: Q3 + 1.5*IQR
    “minimum”: Q1 -1.5*IQR

```python
fresher = np.round(np.random.uniform(3.5, 6.5, size = 10),1)
lead = np.round(np.random.uniform(5.5, 12.5, size = 10),1)
manager = np.round(np.random.uniform(8.5, 16.5, size = 10),1)
director = np.round(np.random.uniform(18.25, 34, size = 10),1)
vp = np.round(np.random.uniform(27, 50.5, size = 10),1)
cxo = np.round(np.random.uniform(45.5,75, size = 10),1)
```

```python
plt.figure(figsize=(15,10))
box_plot_data=[fresher,lead,manager,director,vp,cxo]
plt.boxplot(box_plot_data)
plt.show()
```

```python
plt.figure(figsize=(15,10))
box_plot_data=[fresher,lead,manager,director,vp,cxo]
plt.boxplot(box_plot_data,patch_artist=True,
            labels=['fresher','lead','manager','director','vp','cxo'])
plt.show()
```

```python
plt.figure(figsize=(15,10))
box_plot_data=[fresher,lead,manager,director,vp,cxo]

box = plt.boxplot(box_plot_data,patch_artist=True, 
                  labels=['fresher','lead','manager','director','vp','cxo'], 
                  vert=0)

colors = ['darkblue', 'aqua', 'lime', 'black', 'indigo', 'teal']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    
plt.show()
```
